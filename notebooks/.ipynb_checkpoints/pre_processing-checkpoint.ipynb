### **3.4.2 Remove Noisy Text**

```python
# Define a function for removing noisy text
def text_cleaning(text):
    # remove punctuation
    text = text.translate(str.maketrans('','',string.punctuation))
    # remove HTML
    text = re.compile(r'<.*?>').sub(r'',text)
    # remove emoji
    text = re.compile("["u"\U0001F600-\U0001F64F"  # emoticons
                         u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                         u"\U0001F680-\U0001F6FF"  # transport & map symbols
                         u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                         u"\U00002702-\U000027B0"
                         u"\U000024C2-\U0001F251"
                         "]+", flags=re.UNICODE).sub(r'', text)
    # remove URL
    text = re.compile(r'https?://\S+|www\.\S+').sub(r'',text)
    # remove white spaces
    text = re.sub("\s\s+" , " ", text)
    # remove text in the square brackets
    text = re.sub('\[[^]]*\]','', text)
    # remove line symbol
    text = re.sub('\n', '', text)
    # remove words that contain numbers
    text = re.sub('\w*\d\w*', '', text)
    # remove numbers
    text = re.sub(r'\d+', '', text)

    return text
```

```python
# Removing punctuation with definede function
df['text_clean'] = df['text'].apply(lambda x : text_cleaning(x))
df['description_clean'] = df['description'].apply(lambda x : text_cleaning(x))
```

### **3.4.3 Change Repetitive Characters**

```python
# Create a function for cleaning repetitive characters
def cont_rep_char(text):
    tchr = text.group(0) 
    
    if len(tchr) > 1:
        return tchr[0:2] # take max of 2 consecutive letters
    
def unique_char(rep, tweets):
    substitute = re.sub(r'(\w)\1+', rep, tweets)
    return substitute
```

```python
# Use a loop for changing repetitive characters
df['text_clean'] = df['text_clean'].apply(lambda x : unique_char(cont_rep_char, x))
df['description_clean'] = df['description_clean'].apply(lambda x : unique_char(cont_rep_char, x))
```

## **3.7 Prepare Data for Converting**

### **3.7.1 Remove Stopwords and Short Words**

```python
# Use a loop for removing punctuation
df['text_trans'] = df['text_trans'].apply(lambda x : text_cleaning(x))
df['desc_trans'] = df['desc_trans'].apply(lambda x : text_cleaning(x))
```

```python
def stopwords_shortwords(text):
    # filter out stop words
    words = text.split()
    stop_words = set(stopwords.words( 'english' ))
    words = [w for w in words if not w in stop_words]
    # filter out short tokens
    for word in words:
        if word.isalpha():
            words = [word for word in words if len(word) > 1 ]
        else:
            words = [word for word in words]
    return" ".join(words)
```

```python
df['text_trans'] = df['text_trans'].apply(lambda x : stopwords_shortwords(x))
df['desc_trans'] = df['desc_trans'].apply(lambda x : stopwords_shortwords(x))
```

```python
df['text_trans'].head(30)
```

### **3.7.2 Tokenization**

```python
df['text_trans_token'] = df['text_trans'].apply(word_tokenize)
df['desc_trans_token'] = df['desc_trans'].apply(word_tokenize)
```

### **3.7.3 Stemming**

```python
#defining the object for stemming
porter_stemmer = PorterStemmer()

#defining a function for stemming
def stemming(text):
    stem_text = [porter_stemmer.stem(word) for word in text]
    return stem_text
```

```python
# Use defined function for stemming
df['text_trans_token'] = df['text_trans_token'].apply(lambda x: stemming(x))
df['desc_trans_token'] = df['desc_trans_token'].apply(lambda x: stemming(x))
```

## **3.8 Calculating Similarity Between Reviews and Descriptions**

```python
df['similarity'] = np.nan
def identity_tokenizer(text):
    return text
tfidf = TfidfVectorizer(tokenizer=identity_tokenizer, lowercase=False)  

for r in range(len(df)):
  try:
    doc1 = df['text_trans_token'].loc[r]
    doc2 = df['desc_trans_token'].loc[r]

    docs = [doc1, doc2]

    tdidf_vector = tfidf.fit_transform(docs)

    similarities = cosine_similarity(tdidf_vector)

    df.loc[r,'similarity'] = round(similarities[0][1], 3)
  except:
    # The cosine similarity is a number between 0 and 1
    df.loc[r,'similarity'] = 0.001
```
